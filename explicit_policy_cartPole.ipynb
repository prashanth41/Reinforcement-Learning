```python
import numpy as np
import math
import gym
from gym import wrappers
import matplotlib.pyplot as plt
import random
```

```python
environment = gym.make('CartPole-v0')
```

```python
def policy(state):
    
    if state<0:
        action=0
    else:
        action=1
        
    return action

```

```python
episodes=200
totalRewards = []

for i in range(episodes):
    
    observation = environment.reset()
    episodeRewards = 0    
    done = False
    while not done:      
        action = policy(observation[2])            
        observation_, reward, done, info = environment.step(action)
        episodeRewards += reward
        observation = observation_
    totalRewards.append(episodeRewards)
```

```python
plt.plot(totalRewards,'g')
plt.xlabel('episodes')
plt.ylabel('rewards')
plt.show() 
```
