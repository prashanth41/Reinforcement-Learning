```python
import numpy as np
import math
import gym
import matplotlib.pyplot as plt
import random
```

```python
def policy(speed):
    # 0 - backward, 1 - none, 2 - forward
    if speed < 0:
        return 0
    elif speed >= 0:
        return 2
```

```python
environment = gym.make('MountainCar-v0')
```

```python
episodes = 200
rewards = np.zeros(episodes)

for i in range(episodes):
    observation = environment.reset()
    episodeRewards = 0
    done = False
    while not done:
        action = policy(observation[1])
        observation, reward, done, info = environment.step(action)
        episodeRewards += reward
    
    rewards[i] = episodeRewards
```

```python
plt.plot(rewards,'g')
plt.xlabel('episodes')
plt.ylabel('rewards')
plt.show()
```
